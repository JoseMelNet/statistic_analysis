```python
import os

import numpy as np
import pandas as pd
import scipy.stats

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Reading and load DataBase
path = 'C:\\Users\\melen\\Dropbox\\Python\\statistic_analysis\\db'
os.chdir(path)
df = pd.read_csv('bicicletas-compartidas.csv')
df.head()
```

### Variables Categoricas

```python
df.columns
```

```python
y = df['cuartil-ano'].apply(lambda x: 'cuartil' + str(int(x)))
```

### Diagrama de Barras

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuariles del año')
ax.set_ylabel('Frecuencia')
```

### Decorando el gráfico

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuariles del año')
ax.set_ylabel('Frecuencia')

ax.patches[3].set_facecolor('red')
```

### Diagrama de Torta

```python
fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels = y.value_counts().index)
ax.set_title('Diagrama de Pie')
```

### Gráficos con variables númericas

```python
viento = df['viento']
mean_viento = np.mean(viento)
std_viento = np.std(viento)
```

```python
fig, ax = plt.subplots()
ax.hist(viento, bins = 30)  # bins = 30, define las particiones sobre una variable continua
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuencias')

# Agregando una linea en el promedio

plt.axvline(mean_viento, c = 'r', linestyle = '--', label = 'Promedio')
plt.axvline(mean_viento + std_viento, c = 'k', linestyle = '--', label = '+ 1 Desviacion')
plt.axvline(mean_viento - std_viento, c = 'k', linestyle = '--', label = '- 1 Desviacion')

ax.legend()
```
